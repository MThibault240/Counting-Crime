```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in the compiled files
year_2016 = pd.read_csv("Datasets/2016_compiled.csv", index_col=0)
year_2017 = pd.read_csv("Datasets/2017_compiled.csv", index_col=0)
year_2018 = pd.read_csv("Datasets/2018_compiled.csv", index_col=0)
year_2019 = pd.read_csv("Datasets/2019_compiled.csv", index_col=0)
year_2020 = pd.read_csv("Datasets/2020_compiled.csv", index_col=0)
```

```python
year_2016['Year'] = '2016'
year_2016
```

```python
year_2017['Year'] = '2017'
year_2017
```

```python
year_2018['Year'] = '2018'
year_2018
```

```python
year_2019['Year'] = '2019'
year_2019
```

```python
year_2020['Year'] = '2020'
year_2020
```

```python
all_years = pd.concat([year_2016, year_2017, year_2018, year_2019, year_2020], ignore_index=True)
all_years
```

```python
all_years2 = all_years.drop([38, 30, 22, 14, 6])
all_years2
```

```python
plt.axhline(1, color='gray', linestyle='--')
plt.axhline(2, color='gray', linestyle='--')
plt.axhline(3, color='gray', linestyle='--')
plt.axhline(4, color='gray', linestyle='--')
plt.axhline(5, color='gray', linestyle='--')
plt.axhline(6, color='gray', linestyle='--')
plt.axhline(7, color='gray', linestyle='--')
ax = sns.barplot(x = 'Year', y = 'Percent Increase', hue='Offense Type', data=all_years2)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.yticks([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.5])
#ax.bar_label(ax.containers[0])
plt.ylim(0, 7.5)
plt.show()
```
